```python
import pandas as pd
import numpy as np
```

```python
ränta = pd.read_excel('historiska-borantor.xlsx',header=0)
```

```python
ränta.columns = ränta.iloc[5]
ränta
```

```python
ränta = ränta[6:]
ränta
```

```python
ränta.info()
```

```python
ränta.reset_index()
```

```python
cols = [0,2,3,4,6]

ränta.drop(ränta.columns[cols], axis=1, inplace=True)
```

```python
ränta
```

```python
ränta['Ränteändringsdag'] = pd.to_datetime(ränta['Ränteändringsdag'])
```

```python
ränta.info()
```

```python
ränta['Ränteändringsdag'] = pd.to_datetime(ränta["Ränteändringsdag"]).dt.normalize()
```

```python
ränta = ränta[(ränta['Ränteändringsdag'] > '1999-12-29')]
```

```python
ränta.reset_index(inplace=True)
ränta.drop(ränta.columns[0], axis=1,inplace=True)
ränta.head()
```

```python
ränta.head(7)
```

```python
ränta = ränta.groupby(pd.PeriodIndex(ränta['Ränteändringsdag'], freq="M")).mean()
ränta = ränta.set_index('Ränteändringsdag')
```

```python
ränta_2 = ränta.resample('MS').mean()
```

```python
ränta_2
```

```python
ränta_2 = ränta_2.ffill()
```

```python
ränta_2.index.name = None
```

```python
ränta_2
```

```python
ränta_2.columns.unique()
```

```python
ränta_2.to_excel('listrate.xlsx', index=True)
```
